**Step 1**: Load the Netflix dataset into a pandas DataFrame to start the analysis.

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

**Step 3**: Filter the dataset to include only movies released in the 1990s.


```python
# Filter data for movies released in 1990s
movies_1990s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]
print(movies_1990s)
```

**Step 4**: Find the most frequent movie duration in the 1990s.

```python
# Determine the most frequent movie duration in the 1990s
duration = movies_1990s['duration'].mode()[0]
print('The most frequent movie duration in the 1990s is ' + str(duration) + ' minutes')
```

Step 5: Count the number of short action movies (less than 90 minutes) released in the 1990s.

```python
# Get a summary of the unique genres
print(movies_1990s['genre'].unique())
```

```python
# Count number of short action movies less than 90 minutes released in the 1990s
short_movie_count = len(movies_1990s[(movies_1990s['duration'] < 90) & (movies_1990s['genre'] == 'Action')])
print(short_movie_count)
```
